```
dna2dna = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
dna = 'ATCCGAATACGGTTCGGGTA'
motif = 'CGAA'
Motif_dict = {'A':'T', 'C':'G', 'G':'C', 'T':'A'}
def reverse_complement(dna):
  other = ''
  for base in dna:
    other = other + dna2dna[base]
  return other[ : :-1] 
```

```
def mutate(dna, motif):
  if (dna.find(motif))!= -1:
    print("Motif is in DNA")
  else:
    return dna
  inverted_repeat = reverse_complement(motif)
  if (dna.find(inverted_repeat, dna.find(motif)))!= -1:
    print("Reverse Complement is in DNA")
    new_sequence = dna[0:dna.find(motif)]
    new_sequence2 = dna[dna.find(inverted_repeat) + len(inverted_repeat):]
    final_sequence = new_sequence + new_sequence2
    print(final_sequence)
    print(inverted_repeat)
  else:
    return dna


  
mutate('ATCCGAATACGGTTCGGGTA', 'CGAA')
```
