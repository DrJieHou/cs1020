```
# Q2: For convenience, we have already provided an implementation of the transcribe function that can give you a list of three-letter mRNA codons from the DNA sequence. We have begun to define the required translate{} dictionary to convert mRNA codons to three-letter amino acids, and AA_mapping{} dictionary to convert three-letter amino acids to one-letter amino acids
# The two dictionaries were defined according to the following two tables, and you can directly use them for RNA translation to the Protein sequence.
```

```
def transcribe(seq): 
  RNA_seq = seq.replace('T', 'U')
  ### Define location of start codon (AUG)
  start_codon = RNA_seq.find('AUG')
  codons = []
  for i in range(start_codon, len(RNA_seq), 3): 
    codon = RNA_seq[i:i + 3]
    if len(codon)==3:
        codons.append(codon)
  return codons

transcribe(DNA)
```

```
# Q3: Second, using the two dictionaries, you need to write a new function to get the amino acids sequence from the list of mRNA codons using 'for' loop (Note: if codon of RNA refers to 'Stop', the translation process is completed). You can create a new function to solve the question, or you can also add new codes inside the transcribe() function.
```

```
def amino_Sequence(codons_list, translate, AA_mapping):
  retval = ""
  for codon in codons_list:
    if (translate[codon] != 'Stop'):
      retval += (AA_mapping[translate[codon]])
    else:
      break
  return retval  
```
