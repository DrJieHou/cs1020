```
def transcribe(seq): 
  RNA_seq = seq.replace('T', 'U')
  ### Define location of start codon (AUG)
  start_codon = RNA_seq.find('AUG')
  codons = []
  for i in range(start_codon, len(RNA_seq), 3): 
    codon = RNA_seq[i:i + 3]
    if len(codon)==3:
        codons.append(codon)
  return codons
```

```
def amino_acids_finder(seq): 
  RNA_seq = seq.replace('T', 'U')
  ### Define location of start codon (AUG)
  start_codon = RNA_seq.find('AUG')
  codons = []
  amino_acids = ''
  for i in range(start_codon, len(RNA_seq), 3): 
    codon = RNA_seq[i:i + 3]
    for key in translate.keys():
        if codon == key:
          aa3 = translate[codon]
          if aa3 == 'Stop':
              return amino_acids
          aa1 = AA_mapping[aa3]
          amino_acids = amino_acids + aa1
  
```
